```python
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

"""helpful hints

no 16 has ever beaten a 1
no 13-15 has ever gone to elite 8
"""
```

```python
data = pd.read_csv('../output/2013_maxout_tourney_game_predictions.csv')
sns.distplot(data['wpred_mean'])
sns.distplot(data['lpred_mean'])

plt.figure()

sns.distplot(data['wpred_var'])
sns.distplot(data['lpred_var'])

print data[data['lpred_mean'] > 0.7]
print data[data['wpred_mean'] < 0.3]
```
