```python
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../output/2013_maxout_tourney_score_predictions.csv')

print data.keys()

sns.regplot(data['wscore_mean'], data['wscore_truth'])
sns.regplot(data['lscore_mean'], data['lscore_truth'])

plt.figure()

sns.distplot(data['wscore_var'])
sns.distplot(data['lscore_var'])

print data[data['wscore_var'] > data['wscore_var'].max()-2]
print data[data['wscore_var'] < data['wscore_var'].min()+1]
```
